```python
def predict_srtr(row: Series) -> float:
    row_raw = reverse_log_transform_row(row)
    outcome = 0.
    if row['diab'] == 1:
        outcome *= -0.066605837
    if pandas.isna(row['diab']):
        outcome *= -0.066605837

    if row['education'] == 4:
        outcome *= -0.000272643
    if pandas.isna(row['education']):
        outcome *= -0.000272643

    if row['cig_use'] == 'Y':
        outcome *= 0.065904842

    if row['malig'] == 'Y':
        outcome *= 0.02219013

    if row['ethcat'] == 2:
        outcome *= -0.088669408
    if row['ethcat'] == 7:
        outcome *= -0.434641386
    if row['ethcat'] == 1:
        outcome *= 0.016172921
    if pandas.isna(row['ethcat']):
        outcome *= -0.434641386

    if row['hist_oth_drug_don'] == 'N':
        outcome *= 0.003011283

    if row['hist_cig_don'] == 'Y':
        outcome *= 0.000498645

    if row_raw['diag'] == 1201 or row_raw['diag'] == 1208:
        outcome *= -0.11598044
    if pandas.isna(row_raw['diag']):
        outcome *= -0.11598044

    if row_raw['age_don'] > 55:
        outcome *= 0.022276125
    if row_raw['age_don'] > 50:
        outcome *= 0.024025803
    if row_raw['age_don'] > 40:
        outcome *= 0.024025803
    if row_raw['age_don'] > 20:
        outcome *= 0.001105558
    if row_raw['age_don'] > 15:
        outcome *= 0.0000457


    if row['hist_hypertens_don'] == 'Y':
        outcome *= 0.000140302

    if row_raw['bmi_don_calc'] < 22:
        outcome *= -0.013675144
    if pandas.isna(row_raw['bmi_don_calc']):
        outcome *= -0.054750638


    if row_raw['bun_don'] < 10:
        outcome *= 0.006214231
    if row_raw['bun_don'] < 15:
        outcome *= 0.030163193
    if row_raw['bun_don'] < 25:
        outcome *= 0.000675563

    #TODO: We have values 1-4, mapped to the parameters from excel file
    # Cerebrovascular/Stroke
    if row['cod_cad_don'] == 2:
        outcome *= 0.009943402

    if row['gender_don'] == 'F':
        outcome *= 0.131377454

    if row['donor insulin'] == 'Y':
        outcome *= -0.243727342
    if pandas.isna(row['donor insulin']):
        outcome *= -0.243727342

    if row['ethcat_don'] == 9:
        outcome *= -0.005654728
    if row['ethcat_don'] == 7:
        outcome *= -0.022069858
    if row['ethcat_don'] == 1:
        outcome *= 0.007700082
    if pandas.isna(row['ethcat_don']):
        outcome *= -0.022069858

    if row_raw['height ratio'] > 1.08:
        outcome *= 0.123483738
    if row_raw['height ratio'] > 1.06:
        outcome *= 0.177597673
    if row_raw['height ratio'] > 0.98:
        outcome *= 0.177597673

    if row_raw['weight ratio'] < 0.7:
        outcome *= -0.676215593
    if row_raw['weight ratio'] > 1.6:
        outcome *= -0.616706456
    if row_raw['weight ratio'] > 1.3:
        outcome *= -0.333586524
    if pandas.isna(row_raw['weight ratio']):
        outcome *= -0.32244479

    if row_raw['creat_don'] < 0.8:
        outcome *= -0.0000895
    if pandas.isna(row_raw['creat_don']):
        outcome *= -0.003848936

    if row_raw['age'] < 70:
        outcome *= -0.0048657
    if row_raw['age'] > 40:
        outcome *= 0.011148729
    if pandas.isna(row_raw['age']):
        outcome *= -0.240346908

    if row_raw['bmi_calc'] > 36:
        outcome *= 0.156816331
    if row_raw['bmi_calc'] > 34:
        outcome *= 0.015018691
    if row_raw['bmi_calc'] > 30:
        outcome *= 0.002490638

    if row_raw['hemo_co_tcr'] > 3.5:
        outcome *= -0.000561188
    if pandas.isna(row_raw['hemo_co_tcr']):
        outcome *= -0.002749823

    if row_raw['most_rcnt_creat'] < 0.7:
        outcome *= 1.751207911
    if row_raw['most_rcnt_creat'] > 1.5:
        outcome *= 0.161043865
    if row_raw['most_rcnt_creat'] > 1:
        outcome *= 0.355230656
    if row_raw['most_rcnt_creat'] > 0.9:
        outcome *= 0.004575029

    if row_raw['dial_after_list'] == 'Y':
        outcome *= 0.484377854

    if row_raw['ecmo_trr'] == 1:
        outcome *= 0.350887035

    # TODO ischemic time, correct variable?
    if row_raw['ischtime'] < 120:
        outcome *= -0.00088359
    if row_raw['ischtime'] < 140:
        outcome *= -0.000286891
    if row_raw['ischtime'] < 160:
        outcome *= -0.0000401
    if row_raw['ischtime'] < 300:
        outcome *= 0.000587559

    if pandas.isna(row_raw['ischtime']):
        outcome *= -0.218530997

    if row_raw['med_cond_trr'] != 'ICU':
        outcome *= -0.075556547

    if row_raw['hemo_pcw_tcr'] < 16:
        outcome *= 0.001151401
    if row_raw['hemo_pcw_tcr'] < 18:
        outcome *= 0.000863272
    if row_raw['hemo_pcw_tcr'] > 32:
        outcome *= 0.000288132
    if row_raw['hemo_pcw_tcr'] > 30:
        outcome *= 0.016847899
    if row_raw['hemo_pcw_tcr'] > 28:
        outcome *= 0.013182796

    if row_raw['retransplant'] >= 1:
        outcome *= 0.105895416

    if row_raw['hemo_pa_dia_tcr'] > 32:
        outcome *= 0.003828561
    if row_raw['hemo_pa_dia_tcr'] > 30:
        outcome *= 0.00102561

    if row_raw['hemo_sys_tcr'] < 20:
        outcome *= 0.002339633
    if row_raw['hemo_pa_dia_tcr'] < 25:
        outcome *= 0.003705285
    if row_raw['hemo_pa_dia_tcr'] > 65:
        outcome *= 0.002035943

    if row_raw['rvad ever'] == 1:
        outcome *= 0.025412411

    if row_raw['tah ever'] == 1:
        outcome *= 0.740901872

    if row_raw['tbili'] > 2:
        outcome *= 0.29652491
    if row_raw['tbili'] > 0.4:
        outcome *= 0.199733923

    if row_raw['transfusions'] == 'Y':
        outcome *= 0.289314026

    if row_raw['ventilator_tcr'] == 1:
        outcome *= 0.289314026

    return outcome

y_pred_srtr = X.apply(predict_srtr, axis=1)
```
